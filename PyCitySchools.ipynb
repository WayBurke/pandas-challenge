```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
#====WMB: To be updated to remove PyCitySchools once this code moves=====

school_data_to_load = "PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_total= len(school_data_complete["school_name"].value_counts())
student_total = school_data_complete["student_name"].count()  
budget_total = school_data_complete["budget"].unique().sum()
avg_math_score =school_data_complete["math_score"].mean()
avg_read_score =school_data_complete["reading_score"].mean()
pass_math_score = (school_data_complete["math_score"]>=70).sum()/ student_total*100
pass_read_score = (school_data_complete["reading_score"]>=70).sum()/ student_total*100
pass_overall = (((school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)).sum()/student_total)*100


# Place all of the data found into a summary DataFrame
summary_df = pd.DataFrame({"Total Schools":[school_total],
                              "Total Students": student_total,
                              "Total Budget": budget_total,
                              "Average Math Score": avg_math_score,
                              "Average Reading Score": avg_read_score,
                              "% Passing Math": pass_math_score,
                              "% Passing Reading": pass_read_score,
                              "% Overall Passing": pass_overall})

#Mapped the format for the Total Students and Total Budget to match the desired output

summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)

#DataFrame generated
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Created a copy of the DataFrame
sumSchool_df = school_data_complete.copy()
#sumSchool_df
```

```python
#create variable to hold the school names to be used for the For Loop
##==Updated in VS.. check if works to remove the ['dsfd'] from name field
school_name_df = pd.array(sumSchool_df["school_name"].unique(), dtype=str); school_name_df
```

```python
SummaryTable3 = {"School Name":[], "School Type":[],"Total Students":[ ], "Total School Budget":[],
                "Per Student Budget":[], "Average Math Score":[], "Average Reading Score":[],
                "% Passing Math":[], "% Passing Reading":[], "% Overall Passing":[]}

SummaryTable3_df =pd.DataFrame(SummaryTable3)
count = 0

for i in school_name_df:
    #Get all the data on the current school
    sumSchool_2_df= sumSchool_df.loc[(sumSchool_df["school_name"] == school_name_df[count]), :]
    
    #Get the school Name
    nameSchool = sumSchool_2_df["school_name"].unique()
    
    #Get the school type: Charter vs District
    schoolType = sumSchool_2_df["type"].unique()
    
    #Count Total number of students at that school
    totalStudent =sumSchool_2_df["student_name"].count()
    
    #Retrieve budget for that school
    totalBudget =int(sumSchool_2_df["budget"].unique())
    
    #Calculate budget per student
    budgetPerStud = totalBudget/totalStudent
    
    #Calculate average math score for the school
    avgMath =sumSchool_2_df["math_score"].mean()
    
    #Calculate average reading score for the school
    avgRead =sumSchool_2_df["reading_score"].mean()
    
    #Calculate percentage of students who passed math for the school
    passMath =((sumSchool_2_df["math_score"]>=70).sum()/ totalStudent)*100
    
    #Calculate percentage of students who passed reading for the school
    passRead=((sumSchool_2_df["reading_score"]>=70).sum()/totalStudent)*100
    
    #Calculate percentage of students who passed both math and reading for the school
    passOverall =(((sumSchool_2_df["math_score"]>=70) & (sumSchool_2_df["reading_score"]>=70)).sum()/totalStudent)*100
    
    #Counter to traverse through the school name list/array
    count =count + 1
    
    #Update the dictionary with calculated values
    SummaryTable3_df.loc[len(SummaryTable3_df.index)] = [nameSchool, schoolType, totalStudent, totalBudget, budgetPerStud, avgMath,
                                                        avgRead, passMath, passRead, passOverall] 

SummaryTable3_df.set_index("School Name").sort_values("School Name")
```
