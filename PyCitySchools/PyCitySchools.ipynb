# Module 4: Panda Challenge

```python
# Dependencies and Setup
import pandas as pd

# File to Load 

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Getting a glimpse of the data
school_data_complete.head() 
```

```python
# Calculated fields for the table in order they appear

school_total= len(school_data_complete["school_name"].value_counts())

student_total = school_data_complete["student_name"].count()  

budget_total = school_data_complete["budget"].unique().sum()

avg_math_score =school_data_complete["math_score"].mean()

avg_read_score =school_data_complete["reading_score"].mean()

pass_math_score = (school_data_complete["math_score"]>=70).sum()/ student_total*100

pass_read_score = (school_data_complete["reading_score"]>=70).sum()/ student_total*100

pass_overall = (((school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)).sum()
                /student_total)*100
```

```python
# Place all of the data found into a summary DataFrame
summary_df = pd.DataFrame({"Total Schools":[school_total],
                              "Total Students": student_total,
                              "Total Budget": budget_total,
                              "Average Math Score": avg_math_score,
                              "Average Reading Score": avg_read_score,
                              "% Passing Math": pass_math_score,
                              "% Passing Reading": pass_read_score,
                              "% Overall Passing": pass_overall})

#Mapped the format for the Total Students and Total Budget to match the desired output
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)

#DataFrame generated
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Created a copy of the DataFrame to be used for this section
sumSchool_df = school_data_complete.copy()
```

```python
#create series to hold the school names to be used for the For Loop
school_name_df = sumSchool_df["school_name"].unique(); school_name_df
```

```python
SummaryTable = {"School Name":[], "School Type":[],"Total Students":[ ], "Total School Budget":[],
                "Per Student Budget":[], "Average Math Score":[], "Average Reading Score":[],
                "% Passing Math":[], "% Passing Reading":[], "% Overall Passing":[]}

SummaryTable_df =pd.DataFrame(SummaryTable)
count = 0

for i in school_name_df:
    #Get all the data on the current school
    sumSchool_2_df= sumSchool_df.loc[(sumSchool_df["school_name"] == school_name_df[count]), :]
    
    
    #Get the school Name
    nameSchool =sumSchool_2_df["school_name"].unique()
    
    #Get the school type: Charter vs District
    schoolType = sumSchool_2_df["type"].unique()
    
    
    #Count Total number of students at that school
    totalStudent =sumSchool_2_df["student_name"].count()
    
    #Retrieve budget for that school
    totalBudget =int(sumSchool_2_df["budget"].unique())
    
    #Calculate budget per student
    budgetPerStud = totalBudget/totalStudent
    
    #Calculate average math score for the school
    avgMath =sumSchool_2_df["math_score"].mean()
    
    #Calculate average reading score for the school
    avgRead =sumSchool_2_df["reading_score"].mean()
    
    #Calculate percentage of students who passed math for the school
    passMath =((sumSchool_2_df["math_score"]>=70).sum()/ totalStudent)*100
    
    #Calculate percentage of students who passed reading for the school
    passRead=((sumSchool_2_df["reading_score"]>=70).sum()/totalStudent)*100
    
    #Calculate percentage of students who passed both math and reading for the school
    passOverall =(((sumSchool_2_df["math_score"]>=70) & (sumSchool_2_df["reading_score"]>=70)).sum()/totalStudent)*100
    
    #Counter to traverse through the school name list/array
    count =count + 1
    
    #Update the dictionary with calculated values
    SummaryTable_df.loc[len(SummaryTable_df.index)] = [nameSchool[0], schoolType[0], totalStudent, totalBudget,
                                                       budgetPerStud, avgMath, avgRead, passMath, passRead, passOverall] 

#Removed PerStudent Budget format as it impacts bin creation
#SummaryTable_df["Per Student Budget"] = SummaryTable_df["Per Student Budget"].map("${:,.2f}".format)

SummaryTable_df["Total School Budget"] = SummaryTable_df["Total School Budget"].map("${:,.2f}".format)    
    
SummaryTable_df.set_index("School Name").sort_values("School Name")
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Reduced the table to capture only what is needed
math_score_all = sumSchool_df[["school_name","grade","math_score"]]
math_score_all.head(3)
```

```python
school_grade = math_score_all["grade"].unique() # array to hold the school class grades
count = 0 # count variable to retrieve the elements in school class grade

for i in school_grade:
    if school_grade[count] == "9th":
        math_score_9th = math_score_all.loc[(math_score_all["grade"] == "9th"), :].groupby(["school_name"])
        avg_math_9th = math_score_9th["math_score"].mean()
    elif school_grade[count] == "10th":
        math_score_10th = math_score_all.loc[(math_score_all["grade"] == "10th"), :].groupby(["school_name"])
        avg_math_10th = math_score_10th["math_score"].mean()
    elif school_grade[count] == "11th":
        math_score_11th = math_score_all.loc[(math_score_all["grade"] == "11th"), :].groupby(["school_name"])
        avg_math_11th = math_score_11th["math_score"].mean()
    elif school_grade[count] == "12th":
        math_score_12th = math_score_all.loc[(math_score_all["grade"] == "12th"), :].groupby(["school_name"])
        avg_math_12th = math_score_12th["math_score"].mean()
    count+=1
    
```

```python
#Put the Math Score series into a DataFrame
math_scores_school = pd.DataFrame({"9th":avg_math_9th,
                                  "10th":avg_math_10th,
                                  "11th":avg_math_11th,
                                  "12th":avg_math_12th})

math_scores_school.round(3)
```

  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
